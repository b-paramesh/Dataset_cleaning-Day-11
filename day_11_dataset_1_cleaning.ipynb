Task 2 — Clean the Dataset
Identify and count missing values in each column.
Drop column(s) that are irrelevant for price analysis (example: society or unnamed columns).
Handle missing values in numeric columns such as bath, balcony, etc. (use either dropna() or fillna() depending on your reasoning).
Convert total_sqft to numeric — handle values like "2100 - 2850" by converting them to an average or a single number.
Remove duplicate rows.

```python
df.isnull().value_counts()
```

```python
df.isnull().sum()
```

```python
df.drop(['society'],axis=1,inplace=True,errors='ignore')
df
```

```python
df["bath"].fillna(df["bath"].median(),inplace=True)
df
```

```python
df['balcony'].fillna(0,inplace=True)
```

```python
# total_sqft cleanup
def convert_sqft(x):
  if isinstance(x,str) and '-' in x:
    a,b = x.split('-')
    return (float(a)+float(b))/2
  try:
    return float(x)
  except:
    return None
df['total_sqft']=df['total_sqft'].apply(convert_sqft)
df
```

```python
df.dropna(subset=['total_sqft'],inplace=True)
df.drop_duplicates(inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
avg_house_price=df.groupby('location')['price'].mean()
avg_house_price
```

```python
high_avg_price=df.groupby('location')['price'].mean().sort_values(ascending=False).head(1)
high_avg_price
```

```python
df[['total_sqft','bath','price']].corr()
```

```python
corr = df[['total_sqft', 'bath', 'price']].corr()
print("Correlation Matrix:")
print(corr)

# ---------- 2) Extract Specific Correlation ----------
corr_sqft_price = corr.loc['total_sqft', 'price']
print("\nCorrelation between total_sqft and price:", corr_sqft_price)

# ---------- 3) Interpretation ----------
print("\nINTERPRETATION:")
if corr_sqft_price > 0.7:
    print("• Strong positive correlation → Larger houses generally have higher prices.")
elif corr_sqft_price > 0.4:
    print("• Moderate correlation → Larger houses often cost more, but not always.")
else:
    print("• Weak correlation → Size alone does NOT determine price.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,10))
sns.histplot(df['price'],bins=50,kde=True)
plt.title("Price Distribution")
plt.xlabel("price")
plt.ylabel("friquency")
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['total_sqft'],df['price'],alpha=0.5)
plt.title("Relation ship between the total_sqrt and the price")
plt.xlabel("total_sqft")
plt.ylabel("price")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x=df['bath'], y=df['price'])
plt.title("Bathroom Count vs Price")
plt.xlabel("Number of Bathrooms")
plt.ylabel("Price (Lakhs)")
plt.show()
```

```python
top10=df.groupby('location')['price'].mean().sort_values(ascending=False).head(10)
top10
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top10.values,y=top10.index,palette='viridis')
plt.title("Top 10 Most Expensive Locations")
plt.xlabel("Average Price (Lakhs)")
plt.ylabel("Location")
plt.show()
```
